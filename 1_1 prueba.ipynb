df_velas = pd.DataFrame({
    'open_time': pd.date_range(start='2025-02-15', periods=100, freq='T'),
    'high': np.random.uniform(2750, 2850, 100),
    'low': np.random.uniform(2650, 2750, 100),
})


```python
def calcular_precios_liquidacion(df_trades):
    """Calcula los precios de liquidaciÃ³n para diferentes niveles de apalancamiento."""
    for L in leverage_levels:
        df_trades[f'liq_price_{L}x'] = np.where(
            df_trades['side'] == 'Buy',
            df_trades['price'] * (1 - 1 / L),  # LiquidaciÃ³n para longs
            df_trades['price'] * (1 + 1 / L)   # LiquidaciÃ³n para shorts
        )
    return df_trades

def detectar_liquidaciones(df_trades, df_velas):
    """Detecta si un precio de liquidaciÃ³n fue alcanzado en las velas."""
    df_trades = df_trades.copy()
    df_trades['liquidated'] = False
    df_trades['liquidation_level'] = np.nan
    
    # Convertir open_time de df_velas a timestamp
    df_velas['Date_time'] = pd.to_datetime(df_velas['open_time'], unit='ms')
    df_trades['Date_time'] = pd.to_datetime(df_trades['timestamp'])
    # Redondear df_trades a intervalos de 5 minutos para emparejar con df_velas
    df_trades['Date_time'] = df_trades['Date_time'].dt.floor('5min')
    
    for L in leverage_levels:
        col_liq = f'liq_price_{L}x'
        
        # Cruce con velas: si el precio de liquidaciÃ³n estÃ¡ entre high y low
        merged = df_trades.merge(df_velas, how='left', on='Date_time')
        liquidated_mask = (merged[col_liq] >= merged['Low']) & (merged[col_liq] <= merged['High'])
        
        df_trades.loc[liquidated_mask, 'liquidated'] = True
        df_trades.loc[liquidated_mask, 'liquidation_level'] = L
    
    return df_trades

def generar_heatmap_plotly(df_trades, df_velas):
    """Genera un grÃ¡fico de velas con heatmap de liquidaciones en Plotly."""
    df_liq = df_trades[df_trades['liquidated']]
    
    # GrÃ¡fico de velas
    fig = go.Figure()
    fig.add_trace(go.Candlestick(
        x=df_velas['open_time'],
        open=df_velas['Open'],
        high=df_velas['High'],
        low=df_velas['Low'],
        close=df_velas['Close'],
        name='Velas Japonesas'
    ))
    
    # Heatmap de liquidaciones
    fig.add_trace(go.Histogram2dContour(
        x=df_liq['open_time'],
        y=df_liq['price'],
        colorscale='Viridis',
        contours=dict(showlines=False),
        name='Heatmap de Liquidaciones'
    ))
    
    fig.update_layout(
        title='Heatmap de Liquidaciones y Velas Japonesas',
        xaxis_title='Tiempo',
        yaxis_title='Precio',
        xaxis_rangeslider_visible=False
    )
    
    fig.show()
```

```python
def generar_heatmap(df_trades,df_velas):
    """Genera un heatmap de densidad de liquidaciones ponderado por amount y agrupado en intervalos de 10 en el precio."""
    df_liq = df_trades[df_trades['liquidated']].copy()
    #df_liq=df_trades.copy()
    df_liq['open_time'] = df_liq['Date_time'].astype(int) // 10**6
    print(df_liq['open_time'])
    df_liq['price_bin'] = (df_liq['price'] // 10) * 10  # Agrupar por intervalos de 10
    
    # Agrupar por timestamp y price_bin, sumando el amount
    df_liq_grouped = df_liq.groupby(['open_time', 'price_bin'], as_index=False)['amount'].sum()
    
    # Crear figura
    fig = go.Figure()
    fig.add_trace(go.Candlestick(
        x=df_velas['open_time'],
        open=df_velas['Open'],
        high=df_velas['High'],
        low=df_velas['Low'],
        close=df_velas['Close'],
        name='Velas Japonesas'
    ))
    
    fig.add_trace(go.Histogram2dContour(
        x=df_liq_grouped['open_time'],
        y=df_liq_grouped['price_bin'],
        z=df_liq_grouped['amount'],  # Ponderar por amount agregado
        colorscale='Viridis',
        contours=dict(showlines=False),
        name='Heatmap de Liquidaciones'
    ))
    
    
    fig.update_layout(
        title='Heatmap de Liquidaciones',
        xaxis_title='Tiempo',
        yaxis_title='Precio de LiquidaciÃ³n',
        xaxis=dict(tickformat='%H:%M'),
        yaxis=dict(dtick=10)
    )
    
    fig.show()
    return df_liq_grouped
```

```python
def generar_grafico_barras(df_trades, df_velas):
    """Genera un grÃ¡fico con velas japonesas y lÃ­neas horizontales indicando zonas de liquidaciÃ³n."""
    df_liq = df_trades.copy()
    df_liq['price_bin'] = (df_liq['price'] // 10) * 10  # Agrupar por intervalos de 10
    
    # Calcular la sumatoria de liquidaciones acumuladas hasta el final del open_time
    df_liq['cumulative_amount'] = df_liq.groupby('price_bin')['amount'].cumsum()
    
    # Agrupar por price_bin, sumando el amount acumulado
    df_liq_grouped = df_liq.groupby('price_bin', as_index=False)['cumulative_amount'].max()
    
    # Crear figura
    fig = go.Figure()
    
    # Agregar velas japonesas
    fig.add_trace(go.Candlestick(
        x=df_velas['open_time'],
        open=df_velas['Open'],
        high=df_velas['High'],
        low=df_velas['Low'],
        close=df_velas['Close'],
        name='Velas Japonesas'
    ))
    
    # Agregar lÃ­neas horizontales para cada zona de liquidaciÃ³n
    for _, row in df_liq_grouped.iterrows():
        fig.add_trace(go.Scatter(
            x=[df_velas['open_time'].min(), df_velas['open_time'].max()],
            y=[row['price_bin'], row['price_bin']],
            mode='lines',
            line=dict(width=2, color=f'rgba(255, 0, 0, {row["cumulative_amount"] / df_liq_grouped["cumulative_amount"].max()})'),
            name=f'LiquidaciÃ³n {row["price_bin"]}'
        ))
    
    fig.update_layout(
        title='Zonas de ConcentraciÃ³n de Liquidaciones',
        xaxis_title='Tiempo',
        yaxis_title='Precio de LiquidaciÃ³n',
        yaxis=dict(dtick=10)
    )
    
    fig.show()
```
